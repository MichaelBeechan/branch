这个方法存在一定的风险, 因为它会把 a 字符串里所有的 ‘/home/work’ 都给替换成空内容, 也就是删除. 如果只是自己写小工具, 对输入数据有足够的了解, 也是可以用的. 
# 2. re.sub(‘^/home/work’, ”, a) 
这个需要用到正则表达式, 并且通过锚定的方法, 精确删除字符串开头的 ‘/home/work’, 万无一失

```python
import re  #research
import json
def getCommentCounts(newsurl):
    m = re.search('doc-i(.*).shtml', newsurl)
    newsid = m.group(1)
    comments = requests.get(commentURL.format(newsid))
    jd = json.loads(comments.text.replace('try{jQuery111101457597292487176_1536305761369(','').replace(')}catch(e){}',''))
    return jd['result']['count']['gn:comos-hitesuz7540866:0']['total']
```

```python
import requests
from datetime import datetime
from bs4 import BeautifulSoup

def getNewsDetail(newsurl):
    result = {}
    res = requests.get(newsurl)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, 'html.parser')
    result['title'] = soup.select('.main-title')[0].text   #取得标题
    result['newssource'] = soup.select('.date-source a')[0].text
    timesource = soup.select('.date')[0].contents[0].strip()
    result['dt'] = datetime.strptime(timesource,'%Y年%m月%d日 %H:%M')
    result['article'] = ' '.join([p.text.strip() for p in soup.select('#article p')[:-1]])
    result['comment'] = getCommentCounts(newsurl)
    result['editor'] = soup.select('.show_author')[0].text.lstrip('责任编辑：')
    return result
```

```python
import re  #research
import json
def getCommentCounts(newsurl):
    m = re.search('doc-i(.*).shtml', newsurl)
    newsid = m.group(1)
    comments = requests.get(commentURL.format(newsid))
    jd = json.loads(comments.text.replace('try{jQuery111101457597292487176_1536305761369(','').replace(')}catch(e){}',''))
    return jd['result']['count']['gn:comos-hitesuz7540866:0']['total']
```

```python
import requests
from datetime import datetime
from bs4 import BeautifulSoup

def getNewsDetail(newsurl):
    result = {}
    res = requests.get(newsurl)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, 'html.parser')
    result['title'] = soup.select('.feed-card-content')[0].text   #取得标题
    result['newssource'] = soup.select('.date-source a')[0].text
    timesource = soup.select('.date')[0].contents[0].strip()
    result['dt'] = datetime.strptime(timesource,'%Y年%m月%d日 %H:%M')
    result['article'] = ' '.join([p.text.strip() for p in soup.select('#article p')[:-1]])
    result['comment'] = getCommentCounts(newsurl)
    result['editor'] = soup.select('.show_author')[0].text.lstrip('责任编辑：')
    return result
```

```python
def parserListLinks(url):
    newsdetail = []
    res = requests.get(url)
    jd = json.loads(res.text.replace('try{feedCardJsonpCallback(','').replace(');}catch(e){};',''))
    for ent in jd['result']['data']:
        newsdetail.append(getNewsDetail(ent['url']))
    return newsdetail
```

```python
url = 'http://feed.mix.sina.com.cn/api/roll/get?pageid=143&lid=1528&num=30&versionNumber=1.2.8&page=1&encode=utf-8&callback=feedCardJsonpCallback&_=1536319670045'
#parserListLinks(url)
```

```python
# 提取多页网页中的内容  
news_total = []   #list
for i in range(1, 10):
    newsurl = url.format(i)
    newsary = parserListLinks(newsurl)
    news_total.extend(newsary)
# 后续修改 
```
